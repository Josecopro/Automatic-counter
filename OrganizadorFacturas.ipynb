```python
import pandas as pd 
import os 
import numpy as np

#Python 3.12.7 (seleccionar ese entorno que luego se me olvida)
```

```python
# Obtener archivo xlsx (Excel)
def get_file():
    with os.scandir() as entries:
        for entry in entries:
            if entry.name == "FACTURAS.xlsx":
                return entry.name
        
        return "Archivo no encontrado"

            
print(get_file())
```

```python
# Leer archivo xlsx 
df = pd.read_excel(get_file())

df      
```

```python
Emitido = []
Recibido = []

# Filtrar por columna "Grupo"

for i in range(len(df)):
    if df['Grupo'][i] == 'Emitido':
        Emitido.append(df.iloc[i])
    elif df['Grupo'][i] == 'Recibido':
        Recibido.append(df.iloc[i])
        

```

```python
Recibido = pd.DataFrame(Recibido)
Emitido = pd.DataFrame(Emitido)
```

```python
Recibido.to_excel('Recibido.xlsx')
Emitido.to_excel('Emitido.xlsx')
```
